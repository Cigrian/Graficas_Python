les suele llamar red. Las gráficas pueden ser directas o indirectas y se usan en una gran variedad de contextos.

## Creación de gráficas a partir de una lista de adyacencia

Cada una de las entradas de una lista de adyacencia representa un borde, el cual se encuentra conectado a los demás sobre la misma línea. Solo se requiere definir los bordes una sola vez cuando se trabaja con una gráfica no dirigida. Ejemplo, sí tenemos la lista de adyacencia:

0 1 2 3
1 3 5
3 4

la cual crea una gráfica con nodos llamados 0 a 5, es decir, 6 aristas. Donde el nodo 0 está conectado con el nodo 1, 2 y 3; el nodo 1 se encuentra conectado por el nodo 0, 3 y 5; el nodo 3 se encuentra conectado con el nodo 0 y 4; el nodo 4 solo se encuentra conectado con el nodo 3; y el nodo 4 está conectado al nodo 3 y el nodo 5 está conectado solo con el nodo 1.

Usando la función read_adjlist(filename) se puede cargar una lista de adyacencia, con la cual se puede contruir un gráfico. Se puede usar la opción nodetype=int para que todos los nodos sean tratados como enteros.
para cargar en un grafo un fichero que contenga una lista de adyacencia. La opción nodetype=int indica a networkX que trate los nombres de los nodos como enteros.
