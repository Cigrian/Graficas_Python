# Gráficas

Una gráfica puede entenderse como la unión de varios puntos unos entre otros. A la conexión de diversos puntos se 
les suele llamar red. Las gráficas pueden ser directas o indirectas. LAs gráficas se usan en una gran variedad de contextos.

## Creación de gráficas

Se pueden crear de gráficas en python mediante la libreria networkx de diversas maneras:
- Cargando una carperta que contenga una lista de adyacencia.
- cargando una carpeta que contenga una lista de aristas.
- creada a partir de una matriz de adyacencia.
- a partir de un dataframe de pandas.
- crear un gráfico vacio y añadir los vértices y aristas uno a uno o desde una lista.
